```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eV = np.loadtxt('eV.txt',delimiter='\n')
counts = np.loadtxt('counts.txt')
```

```python
eV2 = np.loadtxt('eV2.txt')
counts2 = np.loadtxt('counts2.txt')
print(eV)
print(eV2)
diff = counts[207:] - counts2
print(diff)
```

```python
plt.plot(eV/10, counts/10)
plt.xlim([-10, 150])
plt.ylim([0, 150000])
plt.plot(eV2/10, counts2)
plt.plot(diff, counts2)
plt.show()
```
