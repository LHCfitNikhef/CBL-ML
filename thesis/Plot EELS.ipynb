```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eV = np.loadtxt('eV.txt',delimiter='\n')
counts = np.loadtxt('counts.txt')
```

```python
eV2 = np.loadtxt('eV2.txt')
counts2 = np.loadtxt('counts2.txt')

diff = (counts[207:]/10 - counts2)

```

```python
plt.plot(eV/10, counts/10, label='Full spectrum')
plt.xlim([-10, 150])
plt.ylim([0, 100000])
plt.plot(eV2/10, counts2, label='Cleaned spectrum')
plt.plot(eV2/10, diff, label='ZLP tail')
plt.legend()
plt.show()

plt.plot(eV[:800]/10, counts[:800]/100000, label='Full spectrum')
plt.plot(eV[800:]/10, counts[800:]/10000, label='Full spectrum')
plt.axvline(x=62, color='black')
plt.xlim([-10, 120])
plt.xlabel('Energy Loss [eV]')
plt.ylim([0, 15])
plt.ylabel('Counts [$\cdot 10^4$]')
plt.show()

plt.plot(eV/10, counts/100000, label='Full spectrum')
#plt.plot(eV2/10, diff/10000, color='red',label='ZLP')
#plt.plot(eV[:208]/10,  counts[:208]/100000, color='red')
plt.xlim([-10, 80])
plt.xlabel('Energy Loss [eV]')
plt.ylim([0, 15])
plt.ylabel('Counts [$\cdot 10^4$]')
plt.legend()

plt.show()
```

```python
plt.xlim([0, 60])
plt.ylim([0, 62000])
plt.plot(eV2/10, counts2, label='Cleaned spectrum')
plt.plot(eV2/10, diff, label='ZLP tail')
plt.title('Low energy loss region')
plt.xlabel('Energy loss [eV]')
plt.ylabel('Intensity [counts]')
plt.legend()
plt.show()

plt.plot(eV2/10, counts2, label='Cleaned spectrum')
plt.xlim([0,10])
plt.ylim([-100, 10000])
plt.xlabel('Energy loss [eV]')
plt.ylabel('Intensity [counts]')
plt.title('ZLP region')
plt.legend()
plt.show()

plt.plot(eV/10, counts/10, label='Full spectrum')
plt.xlim([1,2])
plt.ylim([50000, 200000])
plt.xlabel('Energy loss [eV]')
plt.ylabel('Intensity [counts]')
plt.title('ZLP tail')
plt.legend()
plt.show()

plt.plot(eV/10, counts/10, label='Full spectrum')
plt.xlim([10,30])
plt.ylim([5000, 70000])
plt.xlabel('Energy loss [eV]')
plt.ylabel('Intensity [counts]')
plt.title('Plasmon peaks')
plt.legend()
plt.show()

```
