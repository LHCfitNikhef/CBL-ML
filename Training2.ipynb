```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.preprocessing import PolynomialFeatures
```

```python
N_train = 100
sigma0 = 1
x = np.random.rand(N_train)
sigma = sigma0 * np.random.randn(N_train)
y = 2 * x + sigma

p1 = plt.plot(x,y,'o',color='lightgrey', label='training')

clf = linear_model.LinearRegression()
clf.fit(x[:, np.newaxis], y)
xplot = np.linspace(0,1,N_train*2)
linear_plot=plt.plot(xplot, clf.predict(xplot[:, np.newaxis]), label='Linear')
plt.legend()

poly3 = PolynomialFeatures(degree=3)
X = poly3.fit_transform(x[:,np.newaxis])
clf3 = linear_model.LinearRegression()
clf3.fit(X,y)
Xplot=poly3.fit_transform(xplot[:,np.newaxis])
poly3_plot=plt.plot(xplot, clf3.predict(Xplot), label='Poly 3')



plt.show()
    
```
